```python
import pandas as pd
import fnmatch
import numpy as np
```

```python
crime_df = pd.read_csv("crime_cleaned.csv").drop("Unnamed: 0", axis=1)
```

```python
weather_regions_df = pd.read_csv("weather_regions.csv").drop("Unnamed: 0", axis=1)
```

```python
crime_df
```

```python
weather_regions_df.columns = ["Location", "Date", "Snow_depth", "Region"]
```

```python
weather_regions_df
```

```python
#Remove days from date
weather_regions_df["Date"] = weather_regions_df["Date"].apply(lambda x: x[:-3])
```

```python
def snow_day(x):
    if (x == 17) | (x == 14)|(x == 17) | (x == 18) | (x == 19):
        return 1
    else:
        return 0
 
    
```

```python
weather_regions_df["Snow_days"] = weather_regions_df["Snow_depth"].apply(snow_day)
```

```python
weather_regions_df = weather_regions_df.drop("Snow_depth", axis=1)
```

```python
weather_regions_snow_days = weather_regions_df.groupby(["Location", "Date", "Region"]).sum()
```

```python
weather_regions_snow_days.reset_index(inplace=True)
weather_regions_snow_days
```

```python
final_df = crime_df.merge(weather_regions_snow_days, on=["Region", "Date"])
```

```python
final_df
```

```python
final_df.to_csv("final_dataset.csv")
```
